```python
from sympy import *

HALF = Rational(1,2)
# equations of motion using motor angle theta as variable

r, l, m, I, g = symbols('r, l, m, I, g')
torque = symbols('tau')

t = symbols('t')
theta = Function('theta')(t)


omega = theta.diff(t)

y = Function('y')(theta) # kinematics
v = y.diff(t)

J = y.diff(theta).simplify() #jacobian

# lagrangian
M = I + m*(J**2)
PE = m*g*y

theta_dd_general = (torque - diff(PE,theta) - HALF*diff(M,theta)*omega**2 ) / M
theta_dd_general = theta_dd_general.simplify()

subs_dict = {
    # diff(y, theta): symbols("y'"),
    theta: symbols('theta'),
    diff(theta, t): symbols("\dot{\\theta}"),
}

Eq(symbols('\ddot{\\theta}'), theta_dd_general.subs(subs_dict))
```

```python
tau_max, w_max = symbols('tau_max, w_max')
tau_subs = tau_max * (1-theta.diff(t)/w_max)
```

```python
tsa_eom = theta_dd_general.subs(y, l - sqrt(l**2 - (r*theta)**2)).subs(torque, tau_subs)
Eq(symbols('\ddot{\\theta}'), tsa_eom)
```

```python
repls, reduced = cse(tsa_eom)

for s, e in repls:
    print(f"{s} = {e}")
print("\nreduced:")
display(reduced[0])
```

```python
repls, reduced = cse(tsa_eom.doit().simplify())

for s, e in repls:
    print(f"{s} = {e}")
print("\nreduced:")
display(reduced[0])
```
