```python
from sympy import *

HALF = Rational(1,2)
# equations of motion using motor angle theta as variable

r, l, m, I, g = symbols('r, l, m, I, g')
torque = symbols('tau')

t = symbols('t')
theta = Function('theta')(t)
omega = theta.diff(t)

y = l - sqrt(l**2 - (r*theta)**2) # contraction length as num segments goes to infinity
v = y.diff(t)

# lagrangian
T = HALF*m*v**2 + HALF*I*omega**2
V = m*g*y
L = T - V
L = L.simplify()
L
```

```python
J = y.diff(theta).simplify() #jacobian

M = I + m*(J**2)
M = M.simplify()

C = HALF * (diff(M, theta) + diff(M, theta) - diff(M, theta)) * omega
C = C.simplify()

G = V.diff(theta)
G = G.simplify()

display(M)
display(C)
display(G)
```

```python
theta_dd = (torque - G - C*omega)/M
theta_dd = theta_dd.simplify()
theta_dd
```

```python
euler_lagrange = Eq(diff(diff(L, diff(theta)), t) - diff(L, theta), torque)
euler_lagrange.simplify()
```

```python
second_derivatives = solve(euler_lagrange, diff(diff(theta)))
theta_dd_el = second_derivatives[0].simplify()
theta_dd_el
```

```python
theta_dd_simp = (torque - diff(V,theta) - HALF*diff(M,theta)*omega**2 ) / M
theta_dd_simp = theta_dd_simp.simplify()
theta_dd_simp
```

```python
gamma = Function('gamma')(theta) # mechanical advantage
omega = theta.diff(t)

y = l - gamma*theta # contraction length as num segments goes to infinity
v = y.diff(t)

J = y.diff(theta).simplify() #jacobian

# lagrangian
M = I + m*(J**2)
PE = m*g*y

theta_dd_general = (torque - diff(PE,theta) - HALF*diff(M,theta)*omega**2 ) / M
theta_dd_general = theta_dd_general.simplify()
theta_dd_general
```

```python
theta_dd_general.subs(gamma, sqrt(l**2 - (r*theta)**2)/theta).simplify()
```
